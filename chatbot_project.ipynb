```python
import random

print("🤖 AI Chatbot Started! (type 'bye' to exit)\n")

# Responses dictionary
responses = {
    "hello": ["Hi!", "Hello there!", "Hey!"],
    "how are you": ["I'm good!", "Doing great!", "All fine!"],
    "your name": ["I'm your AI chatbot", "You can call me ChatBot"],
    "what can you do": ["I can chat with you!", "I can answer simple questions"],
    "bye": ["Goodbye!", "See you later!", "Bye!"]
}

while True:
    user = input("You: ").lower()

    if user in responses:
        print("Bot:", random.choice(responses[user]))

        if user == "bye":
            break
    else:
        print("Bot: Sorry, I didn't understand that. Try asking something else.")
```
