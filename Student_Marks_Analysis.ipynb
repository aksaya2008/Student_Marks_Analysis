```python
import pandas as pd
import matplotlib.pyplot as plt

# Student data
data = {
    "Name": ["Arun", "Bala", "Cathy", "David", "Eva"],
    "Marks": [78, 85, 90, 66, 35]
}

df = pd.DataFrame(data)

# Grade function
def grade(mark):
    if mark >= 90:
        return "A"
    elif mark >= 75:
        return "B"
    elif mark >= 50:
        return "C"
    else:
        return "Fail"

# Apply grade
df["Grade"] = df["Marks"].apply(grade)

# Pass/Fail
df["Status"] = df["Marks"].apply(lambda x: "Pass" if x >= 50 else "Fail")

# Show data
print("Student Data:\n")
print(df)

# Statistics
print("\nAverage Marks:", df["Marks"].mean())
print("Highest Marks:", df["Marks"].max())
print("Lowest Marks:", df["Marks"].min())

# Topper
topper = df[df["Marks"] == df["Marks"].max()]
print("\nTopper:\n", topper)

# Graph
colors = ["green" if m >= 50 else "red" for m in df["Marks"]]

plt.bar(df["Name"], df["Marks"], color=colors)
plt.title("Student Marks Analysis")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
print("\nPass Students:", len(df[df["Status"]=="Pass"]))
print("Fail Students:", len(df[df["Status"]=="Fail"]))
```
