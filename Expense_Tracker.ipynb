```python
import pandas as pd
import matplotlib.pyplot as plt

# Empty list to store data
categories = []
amounts = []

print("💰 Expense Tracker")
print("Enter your expenses (type 'done' to finish)\n")

# Input loop
while True:
    category = input("Enter category: ")

    if category.lower() == "done":
        break

    amount = float(input("Enter amount: "))

    categories.append(category)
    amounts.append(amount)

# Create DataFrame
df = pd.DataFrame({
    "Category": categories,
    "Amount": amounts
})

# Display data
print("\n📊 Expense Data:\n")
print(df)

# Total expense
total = df["Amount"].sum()
print("\nTotal Expense:", total)

# Highest expense
max_expense = df.loc[df["Amount"].idxmax()]
print("\nHighest Expense Category:")
print(max_expense)

# Bar chart
plt.bar(df["Category"], df["Amount"])
plt.title("Expense Tracker - Bar Chart")
plt.xlabel("Category")
plt.ylabel("Amount")
plt.show()

# Pie chart
plt.pie(df["Amount"], labels=df["Category"], autopct='%1.1f%%')
plt.title("Expense Distribution")
plt.show()

# Save to CSV
df.to_csv("expenses.csv", index=False)
print("\n✅ Data saved to expenses.csv")
```
